```python
%pip install -qq seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
measurements = {"Height (Inches)": [65.78, 71.52, 69.40, 68.22, 67.79, 68.70, 69.80, 70.01, 67.90, 66.78],
                 "Weight (Pounds)": [112.99, 136.49, 153.03, 142.34, 144.30, 123.30, 141.49, 136.46, 112.37, 120.67]}
measurement_df = pd.DataFrame(data = measurements)

plot = sns.lmplot(measurement_df, x = "Height (Inches)", y = "Weight (Pounds)")
plot.set(title = "Linear Regression of Weight on Height")
```

```python
import numpy as np

x = measurement_df["Height (Inches)"]
y = measurement_df["Weight (Pounds)"]

slope, intercept = np.polyfit(x, y, 1)

print("Slope:", slope)
print("Intercept:", intercept)
```
